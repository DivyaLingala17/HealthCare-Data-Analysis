```python
df.rename(columns={"Patient ID":"Patient_Id",
                   "Date of Birth":"Date_of_Birth",
                   "Medical Condition":"Medical_Condition",
                   "Doctor's Notes":"Doctor's_Notes",
                   "Admit Date":"Admit_Date",
                   "Discharge Date":"Discharge_Date",
                   "Bill Amount":"Bill_Amount",
                   "Age at Admission ":"Age_at_Admission",
                   "Present Age":"Present_Age",
                   "Length of Stay":"Length_of_Stay",
                   "Previous Discharge Date":"Previous_Discharge_Date",
                   "Is Readmission":"Readmission",
                   "Billing Category":"Billing_Category",
                   "Day of Week for Admission":"Day_of_Week_for_Admission",
                   "Weekend Admission flag":"Weekend_Admission_flag",
                   "Bill Per Day":"Bill_Per_Day",
                   "Age Group":"Age_Group",
                   "Lenth of Stay Category":"Lenth_of_Stay_Category"}, inplace=True)
```

```python
from datetime import datetime
df["Patient_Id"]=df["Patient_Id"].astype(str)
df["Name"]=df["Name"].astype(str)
df["Gender"]=df["Gender"].astype(str)
df["Medical_Condition"]=df["Medical_Condition"].astype(str)
df["Treatments"]=df["Treatments"].astype(str)
df["Doctor's_Notes"]=df["Doctor's_Notes"].astype(str)
df["Bill_Amount"]=df["Bill_Amount"].astype(float)
df["Length_of_Stay"]=df["Length_of_Stay"].astype(int)
df["Readmission"]=df["Readmission"].astype(bool) # Changed to boolean
df["Billing_Category"]=df["Billing_Category"].astype(str)
df["Day_of_Week_for_Admission"]=df["Day_of_Week_for_Admission"].astype(str)
df["Weekend_Admission_flag"]=df["Weekend_Admission_flag"].astype(bool) # Changed to boolean
df["Bill_Per_Day"]=df["Bill_Per_Day"].astype(float)
df["Age_Group"]=df["Age_Group"].astype(str)
df["Lenth_of_Stay_Category"]=df["Lenth_of_Stay_Category"].astype(str)
df["Present_Age"]=pd.to_numeric(df["Present_Age"],errors="coerce")
df["Age_at_Admission"]=pd.to_numeric(df["Age_at_Admission"], errors="coerce") # Corrected column name
df["Date_of_Birth"]=pd.to_datetime(df["Date_of_Birth"], errors="coerce",dayfirst=True)
df["Admit_Date"]=pd.to_datetime(df["Admit_Date"], errors="coerce",dayfirst=True)
df["Discharge_Date"]=pd.to_datetime(df["Discharge_Date"], errors="coerce",dayfirst=True)
df["Previous_Discharge_Date"]=pd.to_datetime(df["Previous_Discharge_Date"], errors="coerce",dayfirst=True)
```

```python
df["No_Previous_Discharge_Date"]=df["Previous_Discharge_Date"].isnull().astype(int)
```

```python
df['Prev_Discharge_Missing_Flag']=df['Previous_Discharge_Date'].isnull().astype(int)
```

```python
df_with_dates=df[df['Prev_Discharge_Missing_Flag']==0]
df_with_dates['Days_Between']=(df_with_dates['Admit_Date']-df_with_dates['Previous_Discharge_Date']).dt.days
```

```python
before=len(df)
df=df.drop_duplicates(subset=["Patient_Id", "Admit_Date"],keep="first")
removed=before-len(df)
```

```python
#Calculate Length of Stay(LOS)

df['Length_of_Stay']=(df['Discharge_Date']- df['Admit_Date']).dt.days
```

```python
#Create a flag for incomplete cases(missing Discharge_Date)
df["Incomplete_Case"]=df['Discharge_Date'].isnull()

#set LOS to NaN for incomplete cases(missing discharge date)
df.loc[df['Incomplete_Case'],"Length_of_Stay"]=np.nan

#set LOS to NaN for incomplete cases(data entry errors)
df.loc[df['Length_of_Stay']<0,'Length_of_Stay']=np.nan
```

```python
# create Age Group

bins = [-np.inf, 18, 40, 60, np.inf]
labels=["0-18", "19-40", "41-60", "61+"]
#if Age in df.columns:
df["Age_Group"]=pd.cut(df["Age_at_Admission"], bins=bins, labels=labels)
```

```python
#Extract Admission Month

df['Admit_Month']=df['Admit_Date'].dt.month_name()

#Year_Month Period
df['Admit_YearMonth']=df['Admit_Date'].dt.to_period('M')
```
